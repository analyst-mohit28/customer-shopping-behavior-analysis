```python
import pandas as pd
df= pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

### KPI's

```python
# Total Revenue
total_revenue = df['Purchase Amount (USD)'].sum()
total_revenue
```

```python
# Average Order Value(AOV)
aov = df['Purchase Amount (USD)'].mean()
aov
```

```python
# Total Customer
total_customer = df['Customer ID'].nunique()
total_customer
```

```python
# Revenue By Gender
df.groupby('Gender')['Purchase Amount (USD)'].sum()
```

```python
# Discount Dependency Rate
df['Discount Applied'].value_counts(normalize=True)*100
```

```python
# Revenue by Subscription
df.groupby('Subscription Status')['Purchase Amount (USD)'].sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a new column
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Purchase Frequency Days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```
